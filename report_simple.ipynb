```python
# Import for analyzing and reading data
import numpy as np
import pandas as pd
import sklearn
import scipy as sp
import scipy.stats as stats
import os

# Import the visualization code
import visuals as vs
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Pretty display for notebooks
%matplotlib inline
from IPython.display import display
```

```python
def get_stats(csv):
    data = pd.read_csv(os.path.join("logs", csv))
    print "Total succesful runs: {}/{}".format(data['success'].sum(), len(data))
    print "Rate of Reliability: {}".format(data['success'].sum() * 1.0 / len(data))
```

## No Learning

```python
# Load the 'sim_no-learning' log file from the initial simulation results
file = 'sim_no-learning.csv'
vs.plot_trials(file)
print_stats(file)
```
